```scala
import chisel3._
import chisel3.util._
import chisel3.tester._
import chisel3.tester.RawTester.test
import chisel3.experimental._
import chisel3.internal.firrtl.KnownBinaryPoint
```

```scala
class XDistSearchStruct(MB_BUFFER_WIDTH: Int) extends Bundle {
    val iRows   = UInt(32.W)
    val iYStride = UInt(32.W)
    val pYSearch = Vec(MB_BUFFER_WIDTH, SInt(16.W))
    val pYOrg = Vec(MB_BUFFER_WIDTH, SInt(16.W))
}
```

```scala
class XDistortionxGetSADx(MB_BUFFER_WIDTH: Int) extends Module {
    val io = IO(new Bundle {
        val pcDSS = Input(new XDistSearchStruct(4))
        val out = Output(UInt(32.W))
    })

    val pucCur = io.pcDSS.pYSearch
    val pucOrg = io.pcDSS.pYOrg
    val iStride = io.pcDSS.iYStride
    val iRows = io.pcDSS.iRows

    val uiSum = RegInit(0.U(32.W))

    when(iRows =/= 0.U) {
        uiSum := uiSum + (pucOrg, pucCur).zipped.map((a, b) => Abs(a - b)).reduce(_ + _)
        pucOrg := pucOrg + MB_BUFFER_WIDTH
        pucCur := pucCur + iStride
        iRows := iRows - 1.U
    }

    io.out := uiSum
}
```
